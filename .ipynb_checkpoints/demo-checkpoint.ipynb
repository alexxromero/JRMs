```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
import multiprocessing as mp
import pippa
from pippa import JRM                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                       
```

```python
f = h5py.File('../JDM/datasets/dataset_PtNormed.h5', 'r')
y = np.array(f['target'])
threeM = np.array(f['Constituents']['threeM'])
Nsubs_beta05 = np.array(f['Nsubs/Nsubs_beta05'])
Nsubs_beta10 = np.array(f['Nsubs/Nsubs_beta10'])
Nsubs_beta20 = np.array(f['Nsubs/Nsubs_beta20'])
mass = np.array(f['jet_Mass'])
pT = np.array(f['jet_PT'])
print("Keys: ")
print(f.keys())
f.close()

print("Number of jets: ", y.shape)
```

```python
njobs = mp.cpu_count()
print("Number of cores: ", njobs)
kappa = 0.5
beta = 0.5

def JRM_C2(jet):
    return JRM(jet, 2, kappa=kappa, beta=beta, jet_radius=1.2, optimizer='grid')

with mp.Pool(njobs) as pool:
    results_C2 = pool.map(JRM_C2, threeM)
results_C2 = np.array(results_C2)
```

```python
def JRM_lbfgsvb_C2(jet):
    return JRM(jet, 2, kappa=kappa, beta=beta, jet_radius=1.2, optimizer='L-BFGS-B')

with mp.Pool(njobs) as pool:
    results_C2_lbfgsvb = pool.map(JRM_lbfgsvb_C2, threeM)
results_C2_lbfgsvb = np.array(results_C2_lbfgsvb)
```

```python
def JRM_C3(jet):
    return JRM(jet, 3, kappa=kappa, beta=beta, jet_radius=1.2, optimizer='grid')

with mp.Pool(njobs) as pool:
    results_C3 = pool.map(JRM_C3, threeM)
results_C3 = np.array(results_C3)
```

```python
def JRM_C6(jet):
    return JRM(jet, 6, kappa=kappa, beta=beta, jet_radius=1.2, optimizer='grid')

with mp.Pool(njobs) as pool:
    results_C6 = pool.map(JRM_C6, threeM)
results_C6 = np.array(results_C6)
```

```python
def JRM_C8(jet):
    return JRM(jet, 8, kappa=kappa, beta=beta, jet_radius=1.2, optimizer='grid')

with mp.Pool(njobs) as pool:
    results_C8 = pool.map(JRM_C8, threeM)
results_C8 = np.array(results_C8)
```

```python
f = h5py.File('../JDM/datasets/Cn_optimized.h5', 'r')
C2_old = np.array(f['k05/C2'])
C3_old = np.array(f['k05/C3'])
C4_old = np.array(f['k05/C4'])
C6_old = np.array(f['k05/C6'])
C8_old = np.array(f['k05/C8'])
f.close()
```

```python
print(np.isclose(results_C2, C2_old[:, 0]).all())
print(np.isclose(results_C3, C3_old[:, 0]).all())
print(np.isclose(results_C4, C4_old[:, 0]).all())
print(np.isclose(results_C6, C6_old[:, 0]).all())
print(np.isclose(results_C8, C8_old[:, 0]).all())
```
